```python




import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

file_path = "Civil_Engineering_Regression_Dataset (1).csv"
df = pd.read_csv(file_path)

X = df[['Building_Height']].values  
y = df['Construction_Cost'].values  
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]  
intercept = model.intercept_ 

y_pred = model.predict(X)

r2 = r2_score(y, y_pred) 
mse = mean_squared_error(y, y_pred)  

print(f"Regression Equation: Construction Cost = {slope:.2f} * Building Height + {intercept:.2f}")
print(f"R-squared: {r2:.3f}")
print(f"Mean Squared Error: {mse:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Building_Height'], y=df['Construction_Cost'], label='Actual Data')
plt.plot(df['Building_Height'], y_pred, color='red', label='Regression Line')
plt.xlabel('Building Height')
plt.ylabel('Construction Cost')
plt.title('Simple Linear Regression: Building Height vs Construction Cost')
plt.legend()
plt.show()
```
